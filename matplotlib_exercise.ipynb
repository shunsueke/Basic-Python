## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_temprature = pd.read_csv('data/hist_data.csv')

plt.figure(figsize=(5,6))
b =30
w = 0.5

x1 = df_temprature["aデータ"]
x2 = df_temprature["bデータ"] 


plt.hist(x1,color="orange",label="aデータ",bins = b, alpha=w)
plt.hist(x2,color="green",label="bデータ",bins = b,alpha=w)

plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.legend(loc = "best")
plt.grid(False)
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_length_width = pd.read_csv("data/scatter_data.csv")

plt.figure(figsize=(8,6))

df_class0 = df_length_width[df_length_width["class"] == 0]
df_class1 = df_length_width[df_length_width["class"] == 1]
df_class2 = df_length_width[df_length_width["class"] == 2]

plt.scatter(df_class0["length (cm)"],df_class0["width (cm)"],color="red",label="class0")
plt.scatter(df_class1["length (cm)"],df_class1["width (cm)"],color="blue",label="class1")
plt.scatter(df_class2["length (cm)"],df_class2["width (cm)"],color="green",label="class2")

plt.title("クラス０～２のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.legend(loc = "best")
plt.grid(True)
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line = pd.read_csv("data/line_data.csv")

plt.figure(figsize=(6,4))

plt.scatter(df_line["期間"],df_line["件数"],color= "blue")
plt.plot(df_line["期間"],df_line["件数"],color="blue")

plt.xticks(rotation = 30)
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("費用（円）")
plt.grid(True)
```

```python
df_line = pd.read_csv("data/line_data.csv")

plt.figure(figsize=(6,4))
w = 0.5

x = df_line["期間"]
y = df_line["費用"]

plt.bar(x,y,align = "center",width = w)

plt.xticks(x,["2016下半期","2017上半期","2017下半期","2018上半期","2018半期","2019上半期","2019下半期","2020上半期"],rotation = 30)
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("費用（円）")
plt.grid(True)
```
